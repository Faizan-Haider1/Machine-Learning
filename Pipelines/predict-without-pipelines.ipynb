```python

import numpy as np # linear algebra
import pickle
```

```python

# Load using the found path
ohe_sex = pickle.load(open('/kaggle/input/datasets/faizanhaider01/models/ohe_sex.pkl', 'rb'))
ohe_embarked = pickle.load(open('/kaggle/input/datasets/faizanhaider01/models/ohe_embarked.pkl', 'rb'))
clf = pickle.load(open('/kaggle/input/datasets/faizanhaider01/models/clf.pkl', 'rb'))
```

```python
#pclass	/sex	/age	/sibsp	/parch	/fare	/embarked
test_input = np.array([2,'male',31.0,0,0,10.5,'S'],dtype=object).reshape(1,7)
```

```python
test_input
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_sex #As male so 0,1 for female 1,0
```

```python
test_input_embarked = ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_embarked
```

```python
test_input_age = test_input[:,2].reshape(1,1)
```

```python
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]],test_input_age,test_input_sex,test_input_embarked),axis=1)
```
