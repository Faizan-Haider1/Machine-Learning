```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("/kaggle/input/datasets/faizanhaider01/student-academics/students_academics (1).csv")
```

```python
df.sample(4)
```

```python
df.isnull().mean()*100
```

```python
df['Remarks'].value_counts().plot(kind='bar')
```

```python
df['Remarks'].mode()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['Remarks']=='Pass']['Percentage'].plot(kind='kde', ax=ax)

df[df['Remarks'].isnull()]['Percentage'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Remarks with Pass', 'Remarks with NA']
ax.legend(lines, labels, loc='best')

plt.title('Remarks')
```

```python
Temp = df[df['Remarks']=='Pass']['Percentage']
```

```python
df['Remarks'].fillna(value='Pass', inplace=True)
```

```python
df['Remarks'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


Temp.plot(kind='kde', ax=ax)

# distribution of the variable after imputation
df[df['Remarks'] == 'Pass']['Percentage'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Original variable', 'Imputed variable']
ax.legend(lines, labels, loc='best')

# add title
plt.title('Remarks')
```
