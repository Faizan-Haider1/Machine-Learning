# BIVARIATE ANALYSIS AND MULTIVARIATE ANALYSIS:

```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('Titanic Dataset.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

## (Numerical - Numerical):
* Scatterplot

```python
tips.head()
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue = tips['smoker'],style=tips['sex'],size=tips['size'])
```

# (Numerical - Categorical):
* Bar plot
* Box plot
* Dist plot

```python
titanic.sample(5)
```

```python
sns.barplot(x=titanic['pclass'], y=titanic['age'])
```

```python
sns.barplot(x=titanic['pclass'], y=titanic['age'],hue=titanic['sex'])
```

```python
sns.boxplot(x=titanic['sex'],y=titanic['age'])
```

```python
sns.boxplot(x=titanic['sex'],y=titanic['age'],hue=titanic['survived'])
```

```python
sns.displot(titanic[titanic['survived']==0]['age'],hist=False)
sns.displot(titanic[titanic['survived']==1]['age'],hist=False)
```

# (Categorical - Categorical):
* HeatMap
* ClusterMap

```python
titanic.sample(3)
```

```python
pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
titanic.groupby('sex')['survived'].mean()*100
```

```python
(titanic.groupby('sex')['survived'].mean()*100).plot(kind='pie')
```

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

# PAIRPLOT:
It is collection of scatterplot.If you want to do scatterplot for multiple 
numerical col and you have to check the relation of one numerical coloumn with
other it would be difficult for you to do so.To solve this problem we use pair plot

```python
iris.sample(3)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue = 'species')
```
