```python
import http.client
import json

conn = http.client.HTTPSConnection("weatherapi-com.p.rapidapi.com")

headers = {
    'x-rapidapi-key': "261fafa0fdmshd6cb8af824e12a4p16fe70jsn4a99a76d5d52",
    'x-rapidapi-host': "weatherapi-com.p.rapidapi.com"
}

conn.request("GET", "/current.json?q=53.1%2C-0.13", headers=headers)

res = conn.getresponse()
data = res.read()

weather = json.loads(data.decode("utf-8"))

print("City:", weather["location"]["name"])
print("Country:", weather["location"]["country"])
print("Temperature:", weather["current"]["temp_c"], "°C")
print("Condition:", weather["current"]["condition"]["text"])
print("Humidity:", weather["current"]["humidity"], "%")
print("Wind Speed:", weather["current"]["wind_kph"], "kph")
```
