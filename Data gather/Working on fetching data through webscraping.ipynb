```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
    "Connection": "keep-alive"
}
webpage = requests.get("https://webscraper.io/test-sites/e-commerce/ajax", headers=headers).text
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all("p",class_="description card-text")
```

```python
products = soup.find_all("div", class_="col-md-4 col-xl-4 col-lg-4")
```

```python
specs = []
rating = []
caption = []
price = []
title= []
for i in products:
    specs.append(i.find('p',class_="description card-text").text.strip())
    rating.append(i.find("div",class_='ratings').text.strip())
    caption.append(i.find("div",class_='caption').text.strip())
    price.append(i.find("h4",class_="price float-end card-title pull-right").text.strip())
    title.append(i.find('a',class_="title").text.strip())

df=pd.DataFrame(
    {"title":title, "specs":specs, "rating":rating, "price":price}
)
```
